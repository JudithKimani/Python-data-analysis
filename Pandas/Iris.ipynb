```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header = None)
```

```python
iris=pd.DataFrame(df)
iris
```

Create columns for the dataset

```python
iris.columns = ["sepal_length (in cm)", 'sepal_width (in cm)', 'petal_length (in cm)'
                , 'petal_width (in cm)', 'class']
```

```python
iris.head()
```

Is there any missing value in the dataframe?

```python
iris.isnull().sum()
```

Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
iris['petal_length (in cm)'][10:30] = np.nan
```

```python
iris.iloc[10:30]
```

Good, now lets substitute the NaN values to 1.0

```python
iris = iris.replace(np.nan, 1.0)
```

```python
iris.iloc[10:30]
```

Now let's delete the column class

```python
iris = iris.drop(columns = 'class')
```

```python
iris.head()
```

Set the first 3 rows as NaN

```python
iris.iloc[:3] = np.nan
iris
```

Delete the rows that have NaN

```python
iris = iris.dropna()
iris
```

Reset the index so it begins with 0 again

```python
iris.reset_index()
```
