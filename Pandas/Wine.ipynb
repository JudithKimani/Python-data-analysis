```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_table("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data",sep=",",header=None)
```

```python
wine=pd.DataFrame(df)
```

```python
wine.head()
```

Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
wine.drop(wine.columns[[0, 3, 6, 8, 10, 12, 13]], axis = 1, inplace = True)
wine.head()
```

Assign the columns as below:

```python
wine.columns = ["alcohol","malic_acid","alcalinity_of_ash","magnesium","flavanoids","proanthocyanins","hue"]
wine.head()
```

Set the values of the first 3 rows from alcohol as NaN

```python
wine.alcohol.iloc[:3] = np.NAN
```

```python
wine.head()
```

Now set the value of the rows 3 and 4 of magnesium as NaN

```python
wine.magnesium.iloc[2:4] = np.NAN
wine.head()
```

Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
wine.alcohol = wine.alcohol.fillna(10)
wine.magnesium = wine.magnesium.fillna(100)
wine.head()
```

Count the number of missing values

```python
wine.isnull().sum()
```
