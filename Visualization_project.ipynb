Finding outliers

```python
columns_to_check = ['BMI', 'SleepHours', 'WeightInKilograms']

def find_outliers_iqr(data, threshold=1.5):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    outliers = data[((data < lower_bound) | (data > upper_bound))]
    return outliers

for column in columns_to_check:
    outliers_iqr = find_outliers_iqr(data[column])
    print(outliers_iqr)
    print("\n")
```

#For the missing values i intend to drop some, and imput the ones that might affect the visualization if deleted.
In this case we will use mean, mode median and foward fill to imput data

#MODE

```python
Data['GeneralHealth']= Data['GeneralHealth'].fillna(Data['GeneralHealth'].mode()[0])
Data['PhysicalHealthDays']= Data['PhysicalHealthDays'].fillna(Data['PhysicalHealthDays'].mode()[0])
Data['LastCheckupTime']= Data['LastCheckupTime'].fillna(Data['LastCheckupTime'].mode()[0])
Data['PhysicalActivities']= Data['PhysicalActivities'].fillna(Data['PhysicalActivities'].mode()[0])
Data['SmokerStatus']= Data['SmokerStatus'].fillna(Data['SmokerStatus'].mode()[0])
Data['ECigaretteUsage']= Data['ECigaretteUsage'].fillna(Data['ECigaretteUsage'].mode()[0])
Data['ChestScan']= Data['ChestScan'].fillna(Data['ChestScan'].mode()[0])
Data.head()
```

```python
Data.isnull().sum()
```

#MEAN

```python
Data['SleepHours']= Data['SleepHours'].fillna(Data['SleepHours'].mean())
Data['HeightInMeters']= Data['HeightInMeters'].fillna(Data['HeightInMeters'].mean())
Data['WeightInKilograms']= Data['WeightInKilograms'].fillna(Data['WeightInKilograms'].mean())
Data['BMI']= Data['BMI'].fillna(Data['BMI'].mean())
Data.head()
```

#ForwardFill

```python
Data['HadHeartAttack']= Data['HadHeartAttack'].fillna(Data['HadHeartAttack'].fillna(method='ffill'))
Data['HadStroke']= Data['HadStroke'].fillna(Data['HadStroke'].fillna(method='ffill'))
Data['HadDepressiveDisorder']= Data['HadDepressiveDisorder'].fillna(Data['HadDepressiveDisorder'].fillna(method='ffill'))
Data['HadDiabetes']= Data['HadDiabetes'].fillna(Data['HadDiabetes'].fillna(method='ffill'))
Data['RaceEthnicityCategory']= Data['RaceEthnicityCategory'].fillna(Data['RaceEthnicityCategory'].fillna(method='ffill'))
Data.head(3)
```

```python
Data.isnull().sum()
```

#MEDIAN

```python
Data['MentalHealthDays']= Data['MentalHealthDays'].fillna(Data['MentalHealthDays'].median())
```

#Backfill

```python
Data['AlcoholDrinkers']= Data['AlcoholDrinkers'].fillna(Data['AlcoholDrinkers'].fillna(method='bfill'))
Data['HIVTesting']= Data['HIVTesting'].fillna(Data['HIVTesting'].fillna(method='bfill'))
Data['FluVaxLast12']= Data['FluVaxLast12'].fillna(Data['FluVaxLast12'].fillna(method='bfill'))
Data['PneumoVaxEver']= Data['PneumoVaxEver'].fillna(Data['PneumoVaxEver'].fillna(method='bfill'))
Data['AgeCategory']= Data['AgeCategory'].fillna(Data['AgeCategory'].fillna(method='bfill'))
Data.isnull().sum()
```

```python
Data['TetanusLast10Tdap']= Data['TetanusLast10Tdap'].fillna(Data['TetanusLast10Tdap'].mode()[0])
Data['HighRiskLastYear']= Data['HighRiskLastYear'].fillna(Data['HighRiskLastYear'].mode()[0])
Data['CovidPos']= Data['CovidPos'].fillna(Data['CovidPos'].mode()[0])
Data.isnull().sum()
```

#Barplot
Average mean of each state

```python
average_bmi_by_state = Data.groupby('State')['BMI'].mean().nlargest(10).reset_index()

print(average_bmi_by_state)
```

```python
average_bmi_above_25 = average_bmi_by_state[average_bmi_by_state['BMI'] > 25]

# I had to Sort the filtered DataFrame by BMI in descending order
average_bmi_above_25_sorted = average_bmi_above_25.sort_values(by='BMI', ascending=False)


colors = sns.color_palette("turbo", len(average_bmi_above_25_sorted))

# Create the bar plot
plt.figure(figsize=(10, 6))
bars = plt.barh(average_bmi_above_25_sorted.index, average_bmi_above_25_sorted['BMI'], color=colors)
plt.title('Average BMI by State (BMI > 25)')
plt.xlabel('Average BMI')
plt.ylabel('State')
plt.gca().invert_yaxis()  # Invert y-axis to display states from top to bottom

# Set x-axis limits to start from BMI 25
plt.xlim(25, average_bmi_above_25_sorted['BMI'].max() + 1)

# Add mean BMI value annotation above each bar
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.2f}', ha='left', va='center')

plt.tight_layout()
plt.show()
```

A heatmap to visualize the correlation matrix of numerical variables in the dataframe, revealing potential relationships between different health metrics.
#9. Heatmap of Correlation Matrix

```python
numerical_data = Data.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numerical_data.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

This heatmap will allow you to visualize the correlations between different numerical health metrics in your DataFrame, helping you identify potential relationships between them. Positive correlations are indicated by warmer colors (closer to red), while negative correlations are indicated by cooler colors (closer to blue). The intensity of the color reflects the strength of the correlation.
